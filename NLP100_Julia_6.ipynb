```julia
function tab2period(filename, output)
    open(filename, "r") do f
        s=read(f,String)
        open(output, "w") do w
            write(w, replace.(s, ["\t"=>"\",\""])[1])
        end
    end
end
tab2period("newsCorpora.csv", "newsCorpora2.csv")
```

```julia
using Random
using DataFrames
using CSV
data = DataFrame!(CSV.File("newsCorpora.csv", delim="\t",header=["ID", "TITLE", "URL", "PUBLISHER", "CATEGORY", "STORY", "HOSTNAME", "TIMESTAMP"]))

data=data[isequal.(data.PUBLISHER,"Reuters") .| 
    isequal.(data.PUBLISHER, "Huffington Post") .| 
    isequal.(data.PUBLISHER, "Businessweek") .| 
    isequal.(data.PUBLISHER, "Contactmusic.com") .| 
    isequal.(data.PUBLISHER, "Daily Mail"), :]

function partitionTrainTest(data, at)
    n = nrow(data)
    idx = shuffle(1:n)
    train_idx = view(idx, 1:floor(Int, at*n))
    test_idx = view(idx, (floor(Int, at*n)+1):n)
    data[train_idx,:], data[test_idx,:]
end

train, test=partitionTrainTest(data, 0.8)
valid, test = partitionTrainTest(test, 0.5)
trainL=nrow(train);testL=nrow(test)
println("train: $trainL\ttest: $testL")
check8=trainL/(trainL+testL*2)
println(check8)
CSV.write("train.tsv", dlim="\t", writeheader=true)
CSV.write("test.tsv", dlim="\t",writeheader=true)
CSV.write("valid.tsv", dlim="\t", writeheader=true)
```
