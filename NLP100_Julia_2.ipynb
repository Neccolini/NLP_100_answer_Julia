***12. 1列目をcol1.txtに，2列目をcol2.txtに保存***<br>
各行の1列目だけを抜き出したものをcol1.txtに，2列目だけを抜き出したものをcol2.txtとしてファイルに保存せよ．確認にはcutコマンドを用いよ．

```julia
first_rows_file="col1.txt"
second_rows_file="col2.txt"
first_rows=""
second_rows=""
open(filename,"r") do f
    for line in eachline(f)
        s=split(line)
        global first_rows*=s[1]
        global first_rows*="\n"
        global second_rows*=s[2]
        global second_rows*='\n'
    end
end
    
open(first_rows_file,"w") do w1
        print(w1,first_rows)
end
open(second_rows_file,"w") do w2
        print(w2,second_rows)
end
```

```julia
# ;cut -f 1 ./popular-names.txt
```

***13. col1.txtとcol2.txtをマージ***<br>
12で作ったcol1.txtとcol2.txtを結合し，元のファイルの1列目と2列目をタブ区切りで並べたテキストファイルを作成せよ．確認にはpasteコマンドを用いよ．

```julia
output_file="col1_col2_merged.txt"
res=[]
open(first_rows_file, "r") do f1
    for line in eachline(f1)
        push!(res,line)
    end
end
open(second_rows_file,"r") do f2
    i=1
    for line in eachline(f2)
        res[i]*="\t"
        res[i]*=line
        i+=1
    end
end
open(output_file,"w") do w
    for str in res
        println(w,str)
    end
end
```

```julia
# ;paste col1.txt col2.txt 
```

***16. ファイルをN分割する***<br>
自然数Nをコマンドライン引数などの手段で受け取り，入力のファイルを行単位でN分割せよ．同様の処理をsplitコマンドで実現せよ．

```julia
N=Input()
filename="popular-names1.txt"
res=[]
open(filename,"r") do f
    array=readlines(f)
    p=Int(floor(length(array)/N))
    q=Int(length(array)%N)
    idx=1
    for i=1:N
        if i>q 
            push!(res,array[idx:idx+p-1])
            idx+=p
        else 
            push!(res,array[idx:idx+p])
            idx+=p+1
        end
    end
end
length(res)
```

```julia
# ;split -l  278 ./popular-names1.txt 
```
